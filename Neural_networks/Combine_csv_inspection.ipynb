```python
#target correlation matrix
x1 = 'gic'
def smcorr(x1):
    correlation_matrix = dfx.corr()
    k = 12 #number of variables for heatmap
    cols = correlation_matrix.nlargest(k, x1)[x1].index
    cm = np.corrcoef(dfx[cols].values.T)
    sns.set(font_scale=1.25)
    hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
    return plt.show()
smcorr(x1)
```

### Lets calculate a first order difference ---- How fast is the gic changing from one time step to the next

```python
from pandas.plotting import autocorrelation_plot
```

```python
pd.plotting.autocorrelation_plot(dfx["gic"].resample("1h").median())
```

```python
#del dfx['date']
```

```python
#dfx['date'] = dfx.index
```

```python
shift_in_date = 1 # it counts by one the whole time
```

```python
#first order derivative
dfx['lag_1'] = dfx['gic'] - dfx['gic'].shift(1)/(shift_in_date)
```

```python
pd.plotting.autocorrelation_plot(dfx["lag_1"].resample("1h").median())
```

```python
dfx.lag_1.autocorr() 
```
